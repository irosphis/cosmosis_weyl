```python
group = f['sampler']
print(list(group.keys()))
import h5py

def print_hdf5_structure(group, indent=""):
    """
    递归打印 HDF5 文件的组和数据集结构
    """
    for key in group.keys():
        item = group[key]
        if isinstance(item, h5py.Group):
            print(f"{indent}[组] {key}/")
            print_hdf5_structure(item, indent + "  ")
        else: # 如果是数据集 (Dataset)
            print(f"{indent}- {key} (Shape: {item.shape}, Dtype: {item.dtype})")
print_hdf5_structure(f)
```

```python
import h5py
import numpy as np
import os

os.environ["HDF5_USE_FILE_LOCKING"] = "FALSE"

def get_current_status(file_path):
    with h5py.File(file_path, 'r') as f:
        sampler_attrs = f['sampler'].attrs
        
        # 获取壳层的对数似然和对数体积
        log_l = sampler_attrs['shell_log_l']
        log_v = sampler_attrs['shell_log_v']
        
        # 1. 计算已收集壳层的证据贡献 (Z_i = L_i * V_i)
        # 在对数空间操作：log(Z_i) = log_l + log_v
        log_z_shells = log_l + log_v
        
        # 2. 计算总证据 log(Z_total)
        # 使用 logsumexp 保证数值稳定性
        from scipy.special import logsumexp
        log_z_total = logsumexp(log_z_shells)
        
        # 3. 计算当前活跃点集（最后一层）的证据占比
        # 按照嵌套采样定义，f_live = Z_live / Z_total
        log_z_live = log_z_shells[-1]
        f_live = np.exp(log_z_live - log_z_total)
        
        print(f"n_like: {sampler_attrs['n_like']}")
        print(f"explored: {sampler_attrs['explored']}")
        print(f"logZ: {log_z_total:.4f}")
        print(f"f_live: {f_live:.4f}")
        
        return f_live

# 调用
file_path = "../output/des-y3-maglim-2x2pt-polychord.sampler_status copy.hdf5"
if os.path.exists(file_path):
    get_current_status(file_path)
else:
    print("找不到 HDF5 文件，请确认路径。")
```
